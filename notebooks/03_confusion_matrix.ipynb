```python
import matplotlib.pyplot as plt
# General
import pandas as pd
import seaborn as sns
# Torch
import torch
import torchmetrics
```

```python
preds = torch.empty(1000, dtype=torch.long).random_(39)
targets = torch.empty(1000, dtype=torch.long).random_(39)
confusion_matrix = torchmetrics.functional.confusion_matrix(preds, targets, num_classes=39)

df_cm = pd.DataFrame(confusion_matrix.numpy(), index=range(39), columns=range(39))
plt.figure(figsize=(30, 21))
sns.set(font_scale=1.4)
fig_ = sns.heatmap(df_cm, annot=True, cmap='Purples').get_figure()

plt.show(fig_)

df_cm = pd.DataFrame(confusion_matrix.numpy(), index=range(39), columns=range(39))
plt.figure(figsize=(30, 21))
sns.set(font_scale=1.4)
fig_ = sns.barplot(x="tip", y="day", data=df_cm).get_figure()

plt.show(fig_)

# self.logger.experiment.add_figure("Confusion matrix", fig_, self.current_epoch)
```
